```python
import matplotlib.pyplot as plt
```

```python
squares = [1, 4, 9, 16, 25]
```

<h1>We can see that the graph is incorrect as 4 corresponds to 25</h1>


```python


fig, ax = plt.subplots()
ax.plot(squares, linewidth = 3)
ax.set_title("SquareNumbers", fontsize = 24)#title
ax.set_xlabel("value", fontsize =14)
ax.set_ylabel("square of value", fontsize =14)#put the name of y axis

ax.tick_params(axis='both', labelsize=14)
plt.show()


```

<h1>correctint the plot</h1>

```python

squares = [1, 4, 9, 16, 25]

input_values = [1, 2, 3, 4, 5]
```

```python
fig, ax = plt.subplots()
ax.plot(input_values, squares, linewidth = 3)
```
